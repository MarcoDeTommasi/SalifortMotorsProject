as can be seen by this quick analysis, the only variable that contains outliers is the `time_in_company`. I will now look at these values by evaluating the interquantile range. In order to do that I define the function `percentile` that automatically retrieves the values that excede the quartile ranges of [0.25,0.75] and returns the df of outliers.

```python
def percentile(data,col):
    # Calculate 25th percentile 
    percentile25 = data[col].quantile(0.25)
    # Calculate 75th percentile
    percentile75 = data[col].quantile(0.75)
    # Calculate interquartile range
    iqr = percentile75 - percentile25
    # Calculate upper and lower thresholds for outliers
    limit = percentile75 + 1.5 * iqr
    lower_limit = percentile25 - 1.5 * iqr
    no_outliers=len(data[(data[col]>limit) | (data[col]<lower_limit)])
    print(f'Number of outliers, {col} : ', no_outliers)
    print(f'higher limit exceeded: ', len(data[(data[col]>limit)]))
    print(f'lower limit exceeded: ', len( data[(data[col]<lower_limit) ]))
    print(f'iqr: {iqr}, percentile [0.25,0.75]: [{percentile25}, {percentile75}], limit: {limit}, lower limit: {lower_limit}')
    return data[data[col]>=limit]
```

### Logits distribution
at this point, all the assumption listed below about linearity are fullfilled, the only missing check is about the logits of the model.
In logistic regression, the relationship between a predictor variable and the dependent variable does not need to be linear, however, the log-odds (a.k.a., logit) of the dependent variable with respect to the predictor variable should be linear. Here is the formula for calculating log-odds, where _p_ is the probability of response:
<br>
$$
logit(p) = ln(\frac{p}{1-p})
$$
<br>

to do this we first evaluates the value of $p$

```python
# Get the predicted probabilities of the training data
training_probabilities = model.predict_proba(X_train)
training_probabilities
```

in this vector one has in the first value $1-p$ and in the second $p$, so to calcolate the logit(p) is sufficient to evaluate:

```python
# 1. Copy the `X_train` dataframe and assign to `logit_data`
logit_data = X_train.copy()

# 2. Create a new `logit` column in the `logit_data` df
logit_data['logit'] = [np.log(prob[1] / prob[0]) for prob in training_probabilities]
```

```python
# Plot regplot of `activity_days` log-odds
fig, axes= plt.subplots(nrows=2,ncols=2, figsize=(20,15))
axes[0][0]=sns.regplot(data=logit_data, x='number_projects', y='logit',  ax=axes[0][0],  scatter_kws={'s': 2, 'alpha': 0.5})
axes[0][0].set_title(' logits of number of projects', fontsize='14')
axes[1][0]=sns.regplot(data=logit_data, x='satisfaction', y='logit', ax=axes[1][0], scatter_kws={'s': 2, 'alpha': 0.5})
axes[1][0].set_title('logits of satisfaction', fontsize='14')
axes[0][1]=sns.regplot(data=logit_data, x='last_evaluation', y='logit', ax=axes[0][1], scatter_kws={'s': 2, 'alpha': 0.5})
axes[0][1].set_title('logits of last_evaluation', fontsize='14')
axes[1][1]=sns.regplot(data=logit_data, x='salary', y='logit', ax=axes[1][1], scatter_kws={'s': 2, 'alpha': 0.5})
axes[1][1].set_title('logits of salary', fontsize='14')
```

the function   `get_test_scores` takes as input the true $Y_{test}$ data and the predicted values for the model. Then evaluates the parameters of $accuracy$, $F1$, $Recall$ and $Precision$ and stores them into a dataframe

```python
def get_test_scores(model_name:str, preds, y_test_data):
    '''
    Generate a table of test scores.

    In:
    model_name (string): Your choice: how the model will be named in the output table
    preds: numpy array of test predictions
    y_test_data: numpy array of y_test data

    Out:
    table: a pandas df of precision, recall, f1, and accuracy scores for your model
    '''
    accuracy = accuracy_score(y_test_data, preds)
    precision = precision_score(y_test_data, preds)
    recall = recall_score(y_test_data, preds)
    f1 = f1_score(y_test_data, preds)

    table = pd.DataFrame({'model': [model_name],
                        'precision': [precision],
                        'recall': [recall],
                        'F1': [f1],
                        'accuracy': [accuracy]
                        })

    return table
```

# Storing optimal configuration

the function write_pickle and read_pickle are used to read or store parameters configuration to be used in the models.

```python
path = "../"
```

```python
def write_pickle(path, model_object, save_as:str):
    '''
    In: 
        path:         path of folder where you want to save the pickle
        model_object: a model you want to pickle
        save_as:      filename for how you want to save the model

    Out: A call to pickle the model in the folder indicated
    '''    

    with open(path + save_as + '.pickle', 'wb') as to_write:
        pickle.dump(model_object, to_write)
```

```python
def read_pickle(path, saved_model_name:str):
    '''
    In: 
        path:             path to folder where you want to read from
        saved_model_name: filename of pickled model you want to read in

    Out: 
        model: the pickled model 
    '''
    with open(path + saved_model_name + '.pickle', 'rb') as to_read:
        model = pickle.load(to_read)

    return model
```
